```python
# Read all names

years = range(1880, 2018, 1)
data = []
for y in years:
    fn = f'yob{str(y)}.txt'
    df = pd.read_csv(fn, names = ["FirstName","Sex","Count"])
    df['year'] = y
    data.append(df)
    
df = pd.concat(data,ignore_index = True)

df_m = df[df['Sex']== 'M']
df_f = df[df['Sex']== 'F']
```
